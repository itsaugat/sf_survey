Based on the paper, DiCE, only 8 features are selected as follows : 
- employment years (numerical)
- annual income (numerical)
- no. of open credit accounts (numerical)
- credit history in yrs (numerical)
- loan grade (categorical)
- home ownership (categorical)
- purpose (categorical)
- state of residence (categorical)

The target variable is <b>loan_status</b> with the following values:
- 'Fully Paid' -> 1
- 'Charged Off' -> 0

```python
df = df[['emp_length', 'annual_inc', 'open_acc', 'earliest_cr_line', 'grade', 'home_ownership', 'purpose', 'addr_state', 'loan_status']]
```

```python
df.head()
```

```python
# preprocess employment length
df.replace('n/a', np.nan,inplace=True)
df.emp_length.fillna(value=0,inplace=True)
df['emp_length'].replace(to_replace='[^0-9]+', value='', inplace=True, regex=True)
df['emp_length'] = df['emp_length'].astype(int)
```

```python
df.head()
```

```python
df.isnull().values.any()
```

```python
df['earliest_cr_line'].isnull().sum()
```

```python
# remove nan values for earliest_cr_line column
df = df[df['earliest_cr_line'].notnull()]
```

```python
# preprocess earliest_cr_line to get credit history in years
df.earliest_cr_line = pd.to_datetime(df.earliest_cr_line)

dt = datetime(2018, 6, 29)

dttoday = dt.strftime('%Y-%m-%d')

df.earliest_cr_line = df.earliest_cr_line.apply(lambda x: 
                                                int((np.timedelta64((x - pd.Timestamp(dttoday)),'D').astype(int))/-365))
```

```python
# rename earliest_cr_line column to cr_hist_yrs
df.rename(columns = {'earliest_cr_line':'cr_history_yrs'}, inplace = True)
```

```python
# get the target variable
df = df[(df["loan_status"] == "Fully Paid") | (df["loan_status"] == "Charged Off")]
mapping_dictionary = {"loan_status":{ "Fully Paid": 1, "Charged Off": 0}}
df = df.replace(mapping_dictionary)
```
